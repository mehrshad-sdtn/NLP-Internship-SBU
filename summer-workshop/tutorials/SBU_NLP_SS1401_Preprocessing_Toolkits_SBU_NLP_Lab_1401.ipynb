# **Stanza**

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
!pip install stanza
import stanza
```

```python
stanza.download("fa")
```

```python
nlp = stanza.Pipeline("fa")
```

```python
doc = nlp("جلسه کارآموزی پردازش زبان طبیعی در دانشگاه شهید بهشتی برگزار می شود.")
doc
```

**Tokenizer**

```python
Words = [word.text for sentence in doc.sentences for word in sentence.words]
Words
```

**Lemmatizer**

```python
Lemms = [word.lemma for sentence in doc.sentences for word in sentence.words]
Lemms
```

**Comparison**

```python
for sentence in doc.sentences:
  for word in sentence.words:
    print(word.lemma, color.RED + color.BOLD + "<----->", color.END, word.text)
```

# **Parsivar**

```python
!pip install parsivar
```

```python
from parsivar import Tokenizer, Normalizer, FindStems
```

```python
tmp_text = "به گزارش ایسنا سمینار شیمی آلی از امروز ۱۱ شهریور ۱۳۹۶ در دانشگاه علم و صنعت ایران آغاز به کار کرد. این سمینار تا ۱۳ شهریور ادامه می یابد."
```

**Normalizer**

```python
normalizer = Normalizer()
```

```python
normalizer.normalize(tmp_text)
```

```python
normalizer = Normalizer(date_normalizing_needed=True)
```

```python
normalizer.normalize(tmp_text)
```

```python
normalizer = Normalizer(pinglish_conversion_needed=True)
```

```python
normalizer.normalize("raha be madrese miravad.")
```

**Tokenizer**

```python
tokenizer = Tokenizer()
```

Sentence

```python
sents = tokenizer.tokenize_sentences(normalizer.normalize(tmp_text))
sents
```

Word

```python
words = tokenizer.tokenize_words(normalizer.normalize(tmp_text))
words
```

**Stemmer**

```python
stemmer = FindStems()
print(stemmer.convert_to_stem("بیابیم"))
```
